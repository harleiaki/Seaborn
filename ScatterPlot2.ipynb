```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="darkgrid")
tips=sns.load_dataset("tips")
sns.relplot(x="total_bill", y="tip", data=tips);
plt.show()
sns
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", data=tips);
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips);
plt.show()
```

```python
palette = sns.cubehelix_palette(start=-5, rot=15, dark=0.75)
sns.relplot(x="total_bill", y="tip", hue="size", palette=palette, data=tips);
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", size="size", data=tips);
plt.show()
```

```python
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15,200), data=tips);
plt.show()
```

```python
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
```

```python
tips=sns.load_dataset("tips")
ax=sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips)
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill", y="tip", hue="time", style="time", data=tips)
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill", y="tip", hue="day", style="time", data=tips)
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill", y="tip", size="size", data=tips)
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill", y="tip", hue="size", size="size", data=tips)
plt.show()
```

```python
cmap=sns.cubehelix_palette(dark=.3, light=.8,as_cmap=True)
ax=sns.scatterplot(x="total_bill", y="tip", hue="size", size="size", palette=cmap, data=tips)
plt.show()
```

```python
cmap=sns.cubehelix_palette(dark=.3, light=.8,as_cmap=True)
ax=sns.scatterplot(x="total_bill", y="tip", hue="size", size="size", sizes=(20,200), palette=cmap, legend="full", data=tips)
plt.show()
```

```python
cmap=sns.cubehelix_palette(dark=.3, light=.8,as_cmap=True)
ax=sns.scatterplot(x="total_bill", y="tip", hue="day", size="smoker", palette="Set2", data=tips)
plt.show()
```

```python
ax=sns.scatterplot(x="total_bill", y="tip", s=100, color=".2", marker="+", data=tips)
plt.show()
```

```python
iris=sns.load_dataset("iris")
ax = sns.scatterplot(x=iris.sepal_length, y=iris.sepal_width, hue=iris.species, style=iris.species)
plt.show()
```

```python
import numpy as np,pandas as pd; plt.close("all")
index=pd.date_range("1 1 2000", periods=100, freq="m", name="date")
data=np.random.randn(100,4).cumsum(axis=0)
wide_df=pd.DataFrame(data,index,["a", "b", "c", "d"])
ax=sns.scatterplot(data=wide_df)
plt.show()
```
